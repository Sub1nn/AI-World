```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
help(DBSCAN)
```

```python
df = pd.read_csv('../datasets/Mall_Customers.csv')
df.head()
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
X.head()
```

```python
# Scaling
scalar = StandardScaler()
X_scaled = scalar.fit_transform(X)
```

```python
X_scaled
```

```python
# Model
# eps = 0.5
# min_samples = 5
dbcluster = DBSCAN()
dbcluster.fit(X_scaled)
```

```python
# Labels
label = dbcluster.labels_
label
```

```python
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], c=label)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
# Eps = 
# Min Points =
# n + 1, 2 * n
# 3 or 4
```

```python
X_scaled
```

```python
x_df = pd.DataFrame(X_scaled, columns = X.columns)
x_df.head()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
# min_samples = 3
k = 3
```

```python
near_dist = NearestNeighbors(n_neighbors = k)
near_dist.fit(X_scaled)
```

```python
distances, index = near_dist.kneighbors(X_scaled) # k neighbour returns distances and index, so we only need distances keeping index as blank(_)
distances
```

```python
index
```

```python
import numpy as np
```

```python
k_dists = np.sort(distances[:,-1])
```

```python
k_dists
```

```python
plt.plot(k_dists)
plt.xlabel('Data Points sorted by distance')
plt.ylabel('Distance to {k}-th nearest neighbor'.format(k=k))
plt.title('K-distance Graph')
plt.show()
```

```python
db_cluster = DBSCAN(eps=0.3, min_samples=3)
db_cluster.fit(X_scaled)
```

```python
label = db_cluster.labels_
print(label)
```

```python
plt.scatter(x_df['Annual Income (k$)'], x_df['Spending Score (1-100)'], c=label)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering')
plt.show()
```
